```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from imblearn.under_sampling import RandomUnderSampler
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Load the data
complaints = df

# Select relevant features for X and encode categorical variables
X = complaints[['Product', 'Sub-product', 'Issue', 'State', 'Tags', 'Submitted via', 'Company response to consumer', 'Timely response?']]
X = pd.get_dummies(X)

# Encode the target variable 'Consumer disputed?'
le = preprocessing.LabelEncoder()
y = le.fit_transform(complaints['Consumer disputed?'])

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# Check the proportion of disputed complaints in the training dataset
dispute_proportion = y_train.sum() / len(y_train)
print(f"Dispute proportion in training set: {dispute_proportion:.2f}")

if dispute_proportion < 0.30:
    undersampler = RandomUnderSampler(random_state=123)
    X_train, y_train = undersampler.fit_resample(X_train, y_train)
```

```python
# Train the XGBoost Classifier
model_xgb = XGBClassifier(random_state=123)
model_xgb.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model_xgb.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Calculate costs
costs = {
    'TrueNoDispute': 100,
    'FalseDispute': 190,   # Predicted dispute, but there was no dispute
    'FalseNoDispute': 600, # Predicted no dispute, but there was a dispute
    'TrueDispute': 190     # Predicted dispute correctly
}

# Compute the total cost for the base case
base_case_cost = y_test.sum() * 600 + (len(y_test) - y_test.sum()) * 100
print(f"Base case cost: ${base_case_cost:,}")

# Compute the total cost for the model
tn, fp, fn, tp = cm.ravel()
model_cost = (tn * costs['TrueNoDispute'] + 
              fp * costs['FalseDispute'] + 
              fn * costs['FalseNoDispute'] + 
              tp * costs['TrueDispute'])
print(f"Model cost: ${model_cost:,}")
```

```python
import numpy as np

# Get predicted probabilities
y_probs = model_xgb.predict_proba(X_test)[:, 1]  # Probability of being 'disputed'

# Define the costs as per the given cost structure
cost_structure = {
    'TrueNoDispute': 100,
    'FalseDispute': 190,   # Predicted dispute, but there was no dispute
    'FalseNoDispute': 600, # Predicted no dispute, but there was a dispute
    'TrueDispute': 190     # Predicted dispute correctly
}

# Find the optimal threshold
thresholds = np.linspace(0, 1, 100)
costs_outcomes = []  # This list will store the costs for each threshold

for threshold in thresholds:
    # Apply threshold
    y_pred_adjusted = (y_probs > threshold).astype(int)
    
    # Calculate confusion matrix
    cm = confusion_matrix(y_test, y_pred_adjusted)
    tn, fp, fn, tp = cm.ravel()
    
    # Calculate the total cost for this threshold using the cost structure
    total_cost = (tn * cost_structure['TrueNoDispute'] + 
                  fp * cost_structure['FalseDispute'] + 
                  fn * cost_structure['FalseNoDispute'] + 
                  tp * cost_structure['TrueDispute'])
    costs_outcomes.append((threshold, total_cost))

# Find the threshold with the lowest cost
min_cost_threshold, min_cost = min(costs_outcomes, key=lambda x: x[1])
print(f"Minimum cost of ${min_cost:,} achieved at threshold {min_cost_threshold:.2f}")
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix

# Assuming X_train and y_train are already defined and preprocessed
model = XGBClassifier(random_state=123)
model.fit(X_train, y_train)

# Now you can make predictions
y_probs = model.predict_proba(X_test)[:, 1]
```

```python

# Define the cost function
def calculate_total_cost(y_true, y_pred, cost_TP, cost_FP, cost_FN, cost_TN):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    total_cost = (tp * cost_TP) + (fp * cost_FP) + (fn * cost_FN) + (tn * cost_TN)
    return total_cost

# Set the costs according to the confusion matrix
cost_TP = 190  # True positive cost
cost_FP = 190  # False positive cost
cost_FN = 600  # False negative cost
cost_TN = 100  # True negative cost

# Iterate over a range of threshold values to find the one that minimizes total cost
thresholds = np.linspace(0, 1, 101)
min_cost = float('inf')
best_threshold = 0

for threshold in thresholds:
    # Convert probabilities to binary predictions using the threshold
    y_pred_threshold = (y_probs >= threshold).astype(int)
    
    # Calculate total cost for this threshold
    total_cost = calculate_total_cost(y_test, y_pred_threshold, cost_TP, cost_FP, cost_FN, cost_TN)
    
    # Update min_cost and best_threshold if a new minimum is found
    if total_cost < min_cost:
        min_cost = total_cost
        best_threshold = threshold

print(f"The lowest total cost of ${min_cost} is achieved with a threshold of {best_threshold:.2f}")
```
